```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from sklearn.svm import SVC
from sklearn.datasets import make_blobs, make_circles
from sklearn.cluster import SpectralClustering
from sklearn.metrics import normalized_mutual_info_score

from qiskit import BasicAer
from qiskit.circuit.library import ZZFeatureMap, ZFeatureMap, PauliFeatureMap
from qiskit.utils import QuantumInstance, algorithm_globals
from qiskit_machine_learning.algorithms import QSVC
from qiskit_machine_learning.kernels import QuantumKernel
from qiskit_machine_learning.datasets import ad_hoc_data
```

# Simple Example

```python
X=np.array([[0,1],[1,0],[1,0.5],[0.5,1],[0,2],[2,0],[1,1.5],[1.5,1]])
y=np.array([0,1,1,0,0,1,0,1])
```

## ZZFeature Map

```python
feature_map = ZZFeatureMap(feature_dimension=2, reps=2, entanglement='linear')
backend = QuantumInstance(BasicAer.get_backend('qasm_simulator'), shots=8000)
kernel = QuantumKernel(feature_map=feature_map, quantum_instance=backend)
```

### Computing Kernel in SVC

```python
clf = SVC(kernel=kernel.evaluate)
clf.fit(X,y)
```

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap=plt.cm.Paired)
xlim=plt.gca().get_xlim()
ylim=plt.gca().get_ylim()
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)
plt.contour(XX, YY, Z, colors='g', alpha=0.5, levels=[-0.5,0,0.5], linestyles=['--','-','--'])
#plt.gca().scatter(clf.support_vectors_[:,0], clf.support_vectors_[:,1], s=100, linewidth=1, facecolors='none', edgecolors='k')
plt.show()
```

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap=plt.cm.Paired)
plt.imshow(Z, interpolation='nearest',extent=(XX.min(), XX.max(), YY.min(), YY.max()), aspect='auto', origin='lower', cmap=plt.cm.PuOr_r)
```

```python
clf.predict([1.5,1])
```

### Precomputing Kernel Matrix

```python
matrix_train=kernel.evaluate(x_vec=X)

#fig, axs = plt.subplots(1, figsize=(10, 5))
plt.imshow(np.asmatrix(matrix_train), interpolation='nearest', origin='upper', cmap='Blues')
plt.show()
```

```python
matrix_train
```

```python
clf=SVC(kernel='precomputed')
clf.fit(matrix_train,y)
```

### QSVM

```python
qsvc = QSVC(quantum_kernel=kernel)
qsvc.fit(X,y)

plt.scatter(X[:,0], X[:,1], c=y, cmap=plt.cm.Paired)
xlim=plt.gca().get_xlim()
ylim=plt.gca().get_ylim()
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = qsvc.decision_function(xy).reshape(XX.shape)
plt.contour(XX, YY, Z, colors='g', alpha=0.5, levels=[-0.5,0,0.5], linestyles=['--','-','--'])
#plt.gca().scatter(clf.support_vectors_[:,0], clf.support_vectors_[:,1], s=100, linewidth=1, facecolors='none', edgecolors='k')
plt.show()
```

## 4 Different Feature Maps

```python
backend = QuantumInstance(BasicAer.get_backend('qasm_simulator'), shots=1024)
#ZFeature Map
feature_map1 = ZFeatureMap(feature_dimension=2, reps=2)
kernel1 = QuantumKernel(feature_map=feature_map1, quantum_instance=backend)
#Pauli Feature Map 
feature_map2 = ZFeatureMap(feature_dimension=2, reps=4)
kernel2 = QuantumKernel(feature_map=feature_map2, quantum_instance=backend)
#ZFeature Map
feature_map3 = ZZFeatureMap(feature_dimension=2, reps=2, entanglement='circular')
kernel3 = QuantumKernel(feature_map=feature_map3, quantum_instance=backend)
#ZFeature Map
feature_map4 = PauliFeatureMap(feature_dimension=2, reps=2, entanglement='linear')
kernel4 = QuantumKernel(feature_map=feature_map4, quantum_instance=backend)
```

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap=plt.cm.Paired)
xlim=plt.gca().get_xlim()
ylim=plt.gca().get_ylim()
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T

qsvc1 = QSVC(quantum_kernel=kernel1)
qsvc1.fit(X,y)
Z1 = qsvc1.decision_function(xy).reshape(XX.shape)

qsvc2 = QSVC(quantum_kernel=kernel2)
qsvc2.fit(X,y)
Z2 = qsvc2.decision_function(xy).reshape(XX.shape)

qsvc3 = QSVC(quantum_kernel=kernel3)
qsvc3.fit(X,y)
Z3 = qsvc3.decision_function(xy).reshape(XX.shape)

qsvc4 = QSVC(quantum_kernel=kernel4)
qsvc4.fit(X,y)
Z4 = qsvc4.decision_function(xy).reshape(XX.shape)
```

```python
fig,axs=plt.subplots(2,2,figsize=(10,5))
axs[0,0].scatter(X[:,0], X[:,1], c=y, cmap=plt.cm.Paired)
axs[0,0].imshow(Z1, interpolation='nearest',extent=(XX.min(), XX.max(), YY.min(), YY.max()), aspect='auto', origin='lower', cmap=plt.cm.PuOr_r)
axs[0,0].set_title('Z with rep=2')

axs[0,1].scatter(X[:,0], X[:,1], c=y, cmap=plt.cm.Paired)
axs[0,1].imshow(Z2, interpolation='nearest',extent=(XX.min(), XX.max(), YY.min(), YY.max()), aspect='auto', origin='lower', cmap=plt.cm.PuOr_r)
axs[0,1].set_title('Z with rep=4')

axs[1,0].scatter(X[:,0], X[:,1], c=y, cmap=plt.cm.Paired)
axs[1,0].imshow(Z3, interpolation='nearest',extent=(XX.min(), XX.max(), YY.min(), YY.max()), aspect='auto', origin='lower', cmap=plt.cm.PuOr_r)
axs[1,0].set_title('ZZ with circular entanglement')

axs[1,1].scatter(X[:,0], X[:,1], c=y, cmap=plt.cm.Paired)
axs[1,1].imshow(Z4, interpolation='nearest',extent=(XX.min(), XX.max(), YY.min(), YY.max()), aspect='auto', origin='lower', cmap=plt.cm.PuOr_r)
axs[1,1].set_title('Pauli')

for ax in fig.get_axes():
    ax.label_outer()
```

## Computing kernel in SVC

```python
#Creating separable points
X,y=make_blobs(n_samples=30, centers=2, random_state=5)
#Using quantum kernel
feature_map = ZFeatureMap(feature_dimension=2, reps=2)
backend = QuantumInstance(BasicAer.get_backend('qasm_simulator'), shots=1024)
kernel = QuantumKernel(feature_map=feature_map, quantum_instance=backend)
```

```python
#fitting model
clf=SVC(kernel=kernel.evaluate)
clf.fit(X,y)
```

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap=plt.cm.Paired)
#plotting decision function
ax=plt.gca()
xlim=ax.get_xlim()
ylim=ax.get_ylim()

#creating grid for model evaluation
xx = np.linspace(xlim[0], xlim[1], 100)
yy = np.linspace(ylim[0], ylim[1], 100)
XX,YY = np.meshgrid(xx,yy)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

# plot decision boundary and margins
ax.imshow(Z, interpolation='nearest',extent=(XX.min(), XX.max(), YY.min(), YY.max()), aspect='auto', origin='lower', cmap=plt.cm.PuOr_r)
ax.contour(XX, YY, Z, colors='k', alpha=0.5, levels=[-1,0,1], linestyles=['--','-','--'])

# plot support vectors
#ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100, linewidth=1, facecolors='none', edgecolors='k')
plt.show()
```

```python
#splitting data in model training and test data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.25, random_state=11)
```

```python
backend = QuantumInstance(BasicAer.get_backend('qasm_simulator'), shots=8000)

feature_map_1 = ZFeatureMap(feature_dimension=N_dim, reps=2)
kernel_1 = QuantumKernel(feature_map=feature_map_1, quantum_instance=backend)

feature_map_2 = ZFeatureMap(feature_dimension=N_dim, reps=4)
kernel_2 = QuantumKernel(feature_map=feature_map_2, quantum_instance=backend)

feature_map_3 = ZZFeatureMap(feature_dimension=N_dim, reps=2, entanglement='circular')
kernel_3 = QuantumKernel(feature_map=feature_map_3, quantum_instance=backend)

feature_map_4 = PauliFeatureMap(feature_dimension=N_dim, reps=2, entanglement='linear')
kernel_4 = QuantumKernel(feature_map=feature_map_4, quantum_instance=backend)
```
